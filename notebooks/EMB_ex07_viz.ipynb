# Set-up

### Imports

```python
# --- Data handling and visualization ---
import pandas as pd  # Dataframe tools
from tabulate import tabulate  # Pretty printing for dataframes

import seaborn as sns  # Easier plotting tools
import matplotlib.pyplot as plt
%matplotlib inline
```

### Globals

```python
# --- Visualization ---
rc_dict = {
  "savefig.dpi": 900,  # Saved figure dots-per-inch. 600 is "HD"
  "savefig.facecolor": "white",  # This, combined with transparent setting, keeps saved figs from looking like trash on dark backgrounds 
  "savefig.transparent": False,
  "figure.figsize": (14, 10),  # Default (width, height) of figure
}
plt.rcParams.update(rc_dict)
```

## Data preprocessing

```python
synth_n2v_percomponent = pd.read_parquet("../results/dataframes/dataframe-merged_EMB_ex07_DK_20221003.parquet")
synth_n2v_percomponent.index = range(len(synth_n2v_percomponent))

df = synth_n2v_percomponent.astype({
    'N': 'int8',
    'avgk': 'int8',
    'rep': 'int8',
    'mu': 'float32',
    'gamma': 'float32',
    'tau': 'float32',
    'pfi': 'float32',
    'metric': 'category',
})

df.info()
```

# Figures

```python
plt.figure()
sns.relplot(
    data=df, kind="line",
    x="pfi", y="AUROC",
)

plt.title("Performance on LFRs")
plt.ylabel("AUROC")
plt.xlabel("PFI")
# plt.savefig("../results/plots/scatter_auroc-pfi_EMB_ex04-ex06-_DK_20220928.png")
```
