# Preprocessing

### ex26 (Expanded logistic without biases)

```python
# Small datasets
metadata_ex26 = {
    "PROJECT_ID": "EMB_ex26",
    "RESEARCHERS": "DK",
    "CURRENT_VERSION": "v1.0",
    "DATE": "20230222"
}
TAG_ex26 = "{PROJECT_ID}{CURRENT_VERSION}_{RESEARCHERS}_{DATE}".format(**metadata_ex26)

df_ex26 = pd.read_csv(f"../../results/dataframes/dataframe_{TAG_ex26}.csv", index_col="_UID")


# Concatenate larger datasets
metadata_ex26.update({
    "CURRENT_VERSION": "v1.1"
})
TAG_ex26 = "{PROJECT_ID}{CURRENT_VERSION}_{RESEARCHERS}_{DATE}".format(**metadata_ex26)

df_ex26 = pd.concat([df_ex26, pd.read_csv(f"../../results/dataframes/dataframe_{TAG_ex26}.csv", index_col="_UID")])
```

## Concatenating

```python
# Adding identifiers
df_ex23["experiment"] = ["ex23"] * len(df_ex23)
df_ex24["experiment"] = ["ex24"] * len(df_ex24)
df_ex25["experiment"] = ["ex25"] * len(df_ex25)
df_ex26["experiment"] = ["ex26"] * len(df_ex26)
```

```python
# Concatenating across experiments
df = pd.concat([df_ex23, df_ex24, df_ex25, df_ex26])
```

## Pruning

```python
df = df.drop(columns=["intercept", "coef"])
```

## Reindexing and renaming

```python
# Reset index after concatenations
df = df.reset_index()
df = df.drop(columns=["_UID"])
```

```python
# Rename layers
df = df.rename(columns={"l1": "alpha", "l2": "beta"})
```

# Analysis

## N2V distance-based likelihood

```python
# Instantiate figure, subset relevant data
plt.figure()
df_ = df.query("experiment == 'ex23' and theta != 1.0")

# Plotting data
sns.relplot(
    data=df_, kind="line",
    x="theta", y="auroc",
    hue="system",
    marker="o",
    height=6, aspect=1.2
)

# Axes configuration
plt.hlines(0.5, 0.05, 0.9, linestyles="dashed", colors="gray")
plt.xticks(
    [float(np.format_float_scientific(x, precision=1)) for x in set(df_["theta"])],
    rotation=0
)
plt.xlabel("$\\theta$")

plt.yticks(np.linspace(0, 1, endpoint=True, num=10+1))
plt.ylabel("ROC-AUC")


# Figure-level attributes
plt.title("N2V distance-based likelihood reconstruction")

# Save plot
plt.tight_layout()
plt.savefig("../../results/plots/line_auroc-pfi_N2V-distance-likelihood.png", facecolor="white", transparent=False)
```
