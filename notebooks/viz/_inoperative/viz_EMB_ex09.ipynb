# Set-up

## Imports

```python
# --- Standard library ---
from datetime import datetime

# --- Data handling and visualization ---
import pandas as pd  # Dataframe tools
from tabulate import tabulate  # Pretty printing for dataframes

import seaborn as sns  # Easier plotting tools
import matplotlib.pyplot as plt
%matplotlib inline
```

## Globals

```python
today = datetime.today().strftime("%Y%m%d")

# --- Visualization ---
rc_dict = {
  "savefig.dpi": 900,  # Saved figure dots-per-inch. 600 is "HD"
  "savefig.facecolor": "white",  # This, combined with transparent setting, keeps saved figs from looking like trash on dark backgrounds 
  "savefig.transparent": False,
  "figure.figsize": (14, 10),  # Default (width, height) of figure
}
plt.rcParams.update(rc_dict)
```

# Data preprocessing

```python
# Load dataframes
df_S = pd.read_parquet("../../results/dataframes/dataframe_EMB_ex09-Sv1.0_DK_20221010.parquet")
df_R = pd.read_parquet("../../results/dataframes/dataframe_EMB_ex09-Rv1.1_DK_20221011.parquet")
```

# Figures

## ex09-S

```python
plt.figure()
sns.relplot(
    data=df_S, kind="line",
    x="pfi", y="AUROC",
)

plt.title("Performance on LFRs")
plt.ylabel("AUROC")
plt.xlabel("PFI")

plt.savefig(f"../../results/plots/scatter_auroc-pfi_EMB_ex09-S_DK_{today}.png")
```

## ex09-R

```python
plt.figure()
sns.relplot(
    data=df_R, kind="line",
    x="pfi", y="AUROC",
    hue="system",
)

plt.title("Performance on reals")
plt.ylabel("AUROC")
plt.xlabel("PFI")

plt.savefig(f"../../results/plots/scatter_auroc-pfi_EMB_ex09-R_DK_{today}.png")
```

```python
plt.figure()
sns.relplot(
    data=df_R, kind="line",
    x="pfi", y="AUROC",
    hue="dimension", palette="viridis",
    col="system",
)

plt.ylabel("AUROC")
plt.xlabel("PFI")

plt.savefig(f"../../results/plots/scatter_auroc-pfi-dimension_EMB_ex09-R_DK_{today}.png")
```
