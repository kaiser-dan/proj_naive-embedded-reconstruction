```python
metadata_ex21 = {
    "PROJECT_ID": "EMB_ex21",
    "RESEARCHERS": "DK",
    "CURRENT_VERSION": "v3.2.1",
    "DATE": "20230214"
}
TAG_ex21 = "{PROJECT_ID}{CURRENT_VERSION}_{RESEARCHERS}_{DATE}".format(**metadata_ex21)

metadata_ex22 = {
    "PROJECT_ID": "EMB_ex22",
    "RESEARCHERS": "DK",
    "CURRENT_VERSION": "v1.1",
    "DATE": "20230214"
}
TAG_ex22 = "{PROJECT_ID}{CURRENT_VERSION}_{RESEARCHERS}_{DATE}".format(**metadata_ex22)
```

```python
df_ex21 = pd.read_csv(f"../../results/dataframes/dataframe_{TAG_ex21}.csv", index_col="_UID")
df_ex21.head()
```

```python
df_ex22 = pd.read_csv(f"../../results/dataframes/dataframe_{TAG_ex22}.csv", index_col="_UID")
df_ex22.head()
```

# LC-PC-CB-IC

## Small real duplexes

```python
plt.figure()

sns.relplot(
    data=df_ex21, kind="line",
    x="theta", y="auroc",
    hue="system",
    marker="o",
    height=6, aspect=8/6
)

plt.xticks(np.linspace(start=0, stop=0.5, endpoint=True, num=11))
plt.yticks(np.linspace(start=0, stop=1, endpoint=True, num=11))

plt.xlabel("$\\theta$")
plt.ylabel("ROC-AUC")
plt.title("LE-PC-CB-IC small real duplexes")

plt.tight_layout()
plt.savefig(f"../../results/plots/line_auroc-pfi_{TAG_ex21}.png", facecolor="white", transparent=False)
```

## Overlapping Synthetic

```python
plt.figure()

sns.relplot(
    data=df_ex22, kind="line",
    x="theta", y="auroc",
    hue="prop_overlap", palette="coolwarm",
    marker="o",
    height=6, aspect=8/6
)

plt.xticks(np.linspace(start=0, stop=0.5, endpoint=True, num=11))
plt.yticks(np.linspace(start=0.4, stop=1, endpoint=True, num=11))

plt.xlabel("$\\theta$")
plt.ylabel("ROC-AUC")
plt.title("LE-PC-CB-IC overlapping synthetic")

plt.tight_layout()
plt.savefig(f"../../results/plots/line_auroc-pfi_hue-overlap_{TAG_ex22}.png", facecolor="white", transparent=False)
```

```python
plt.figure()

sns.relplot(
    data=df_ex22.query("theta == 0.3"), kind="line",
    x="prop_overlap", y="auroc",
    hue="prop_overlap", palette="Spectral_r", legend=False,
    marker="o",
    height=6, aspect=8/6
)

# plt.xticks(np.linspace(start=0, stop=0.5, endpoint=True, num=11))
# plt.yticks(np.linspace(start=0.4, stop=1, endpoint=True, num=11))

plt.xscale("log")

plt.text(0.08, 0.75, "Fixed $\\theta = 0.3$", fontdict={"size": 12})

plt.xlabel("Proportion Overlap")
plt.ylabel("ROC-AUC")
plt.title("LE-PC-CB-IC overlapping synthetic")

plt.tight_layout()
plt.savefig(f"../../results/plots/scatter_auroc-overlap_{TAG_ex22}.png", facecolor="white", transparent=False)
```

```python
tmp = pd.read_csv(f"../../results/dataframes/dataframe_{TAG_ex21_noPC_noIC}.csv", index_col="_UID")
metadata_ex21_noPC_noIC.update({"CURRENT_VERSION": "v2.1"})
TAG_ex21_noPC_noIC = "{PROJECT_ID}{CURRENT_VERSION}_{RESEARCHERS}_{DATE}".format(**metadata_ex21_noPC_noIC)


df_ex21_noPC_noIC = pd.read_csv(f"../../results/dataframes/dataframe_{TAG_ex21_noPC_noIC}.csv", index_col="_UID")
df_ex21_noPC_noIC = pd.concat([tmp, df_ex21_noPC_noIC])
df_ex21_noPC_noIC = df_ex21_noPC_noIC.reset_index()
```
