```python
import os

from EMB import mplxio, netsci, utils
```

# Prototyping

```python
celegans_raw = mplxio.from_edgelist(os.path.join("..", "data", "edgelists", "multiplex-celegans_l1-1_l2-2.mplx"))
```

```python
all_nodes = sorted(netsci.preprocessing.get_all_nodes(*celegans_raw.values()))
```

```python
current_id_to_new_id = utils.inverse_map(dict(enumerate(all_nodes)))
```

```python
celegans_clean = dict()

for layer_id, layer_graph in celegans_raw.items():
    celegans_clean[layer_id] = netsci.models.make_nodes_contiguous(layer_graph, current_id_to_new_id)
```

# Application

```python
for real_network_fp in os.listdir(os.path.join("..", "..", "..", "data", "edgelists")):
    if "LFR" in real_network_fp or "clean" in real_network_fp:
        continue
    print(real_network_fp)

    mplx_raw = mplxio.from_edgelist(os.path.join("..", "..", "..", "data", "edgelists", real_network_fp))
    all_nodes = sorted(netsci.utils.all_nodes(*mplx_raw.values()))

    current_id_to_new_id = utils.inverse_map(dict(enumerate(all_nodes)))

    mplx_clean = dict()

    for layer_id, layer_graph in mplx_raw.items():
        mplx_clean[layer_id] = netsci.models.make_nodes_contiguous(layer_graph, current_id_to_new_id)

    mplxio.to_edgelist(mplx_clean, os.path.join("..", "..", "..", "data", "edgelists", f"clean-{real_network_fp}"))
```
